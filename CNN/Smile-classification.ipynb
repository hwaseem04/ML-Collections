```python
import torch
from torchvision.datasets import CelebA
from torchvision import transforms
import matplotlib.pyplot as plt
```

```python
image_path = './'
celeb_train = CelebA(
    root = image_path,
    split = 'train',
    target_type = 'attr',
    download = False)
celeb_validation = CelebA(
    root = image_path,
    split = 'valid',
    target_type = 'attr',
    download = False)
celeb_test = CelebA(
    root = image_path,
    split = 'test',
    target_type = 'attr',
    download = False)
```

## Manual augmentation examples
Deterministic
- Cropping at random spot
- Flip horizontal
- Constrast change
- Brightness change
- Center crop and resize

```python
plt.figure(figsize=(16, 8.5))

plt.subplot(2,5,1)
img, attr = celeb_train[-5]
plt.title("Crop at some bounding box", size=15)
plt.imshow(img)

plt.subplot(2,5,6)
crop_img = transforms.functional.crop(img, 50,20,128,128) # x,y, width, height
plt.imshow(crop_img)

plt.subplot(2,5,2)
img, attr = celeb_train[-1]
plt.title("Flip horizontal", size=15)
plt.imshow(img)

plt.subplot(2,5,7)
flip_img = transforms.functional.hflip(img)
plt.imshow(flip_img)

plt.subplot(2,5,3)
img, attr = celeb_train[-2]
plt.title("Adjust contrast", size=15)
plt.imshow(img)

plt.subplot(2,5,8)
const_img = transforms.functional.adjust_contrast(img, contrast_factor=2)
plt.imshow(const_img)

plt.subplot(2,5,4)
img, attr = celeb_train[-3]
plt.title("Adjust brightness", size=15)
plt.imshow(img)

plt.subplot(2,5,9)
bright_img = transforms.functional.adjust_brightness(img, brightness_factor=1.5)
plt.imshow(bright_img)

plt.subplot(2,5,5)
img, attr = celeb_train[-4]
plt.title("Center crop and resize", size=15)
plt.imshow(img)

plt.subplot(2,5,10)
img_center_crop = transforms.functional.center_crop(img, [218*0.6,178*0.6])
img_resize = transforms.functional.resize(img_center_crop, [218,178])
plt.imshow(img_center_crop)

plt.tight_layout()
```
